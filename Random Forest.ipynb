## Data preprocessing

```python
from sklearn import model_selection, preprocessing, linear_model, naive_bayes, metrics
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer 
from sklearn import decomposition, ensemble
import xgboost
import textblob, string
from keras.preprocessing import text, sequence
from keras import layers, models, optimizers
from sklearn.metrics import f1_score
import tensorflow as tf
```

```python
train_x = train['Text']
train_y = train['Label']
test_x = test['Text']
test_y = test['Label']
```

```python
encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
test_y = encoder.fit_transform(test_y)
```

```python
count_vect = CountVectorizer(analyzer='word', token_pattern=r'\w{1,}')
count_vect.fit(train['Text'])
xtrain_count = count_vect.transform(train_x)
xvalid_count = count_vect.transform(test_x)
```

```python
tfidf_vect = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', max_features=5000)
tfidf_vect.fit(train['Text'])
xtrain_tfidf =  tfidf_vect.transform(train_x)
xtest_tfidf =  tfidf_vect.transform(test_x)
```

```python
tfidf_vect_ngram = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', ngram_range=(2,3), max_features=5000)
tfidf_vect_ngram.fit(train['Text'])
xtrain_tfidf_ngram =  tfidf_vect_ngram.transform(train_x)
xvalid_tfidf_ngram =  tfidf_vect_ngram.transform(test_x)
```

## Model

```python
def train_model(classifier, feature_vector_train, label):
  classifier.fit(feature_vector_train, label)
  return classifier
```

```python
def pred_model(classifier, feature_vector_valid, is_neural_net=False):
    predictions = classifier.predict(feature_vector_valid)
    if is_neural_net:
      predictions = predictions.argmax(axis=-1)
    return predictions
```

### Random Forest


```python
# model1 = train_model(ensemble.RandomForestClassifier(), xtrain_tfidf, train_y)
model2 = train_model(linear_model.LogisticRegression(),xtrain_tfidf, train_y)
pred_train2 = pred_model(model2,xtrain_tfidf)
pred_test2 = pred_model(model2,xtest_tfidf)
```

```python
print ("F1-score on training set RF: ", f1_score(train_y, pred_train2,average = 'weighted'))
print ("Accuracy on training set RF: ", metrics.accuracy_score(pred_train2, train_y))
print ("F1-score on testing set RF: ", f1_score(test_y, pred_test2,average = 'weighted'))
print ("Accuracy on testing set RF: ", metrics.accuracy_score(pred_test2, test_y))
```
